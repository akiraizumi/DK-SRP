# 入力データとラベルデータを作成する

```python

# pandas のデータフレームから入力データXとラベルデータYを作る
array = df.values
X = array[:,0:8]
Y = array[:,8]

# ノーマライズ（データの単位が異なるため）
from sklearn.preprocessing import normalize
X=normalize(X)

print("X.shape=",X.shape)
print("Y.shape=",Y.shape)

print("X=",X)
print("Y=",Y)
```
