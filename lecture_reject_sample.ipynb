```
import numpy as np

order=3
np.random.seed(1234)
# データ数と真のパラメータを指定して、真の関数を構築する
N=10
w=[-2,0,2,1,1]
def f(x):
  return np.poly1d(w)(x) 

x=np.linspace(-2,2,100)
y=f(x)


# 描画する
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.ylim(0,3)
plt.xlim(-2,2)
```

```
N=1000
data_x=np.random.uniform(-1.5,1.5,N)
data_y=np.random.uniform(0,2.5,N)
import matplotlib.pyplot as plt
plt.plot(x,y,c="red")
plt.scatter(data_x,data_y)
plt.ylim(0,3)
plt.xlim(-2,2)
```

```
data_x=np.random.uniform(-1.5,1.5,N)
data_y=np.random.uniform(0,2.5,N)
data_thresh=f(data_x)
idx=data_thresh>data_y
accepted_x=data_x[idx]
accepted_y=data_y[idx]
  
import matplotlib.pyplot as plt
plt.plot(x,y,c="red")
plt.scatter(accepted_x,accepted_y)
plt.ylim(0,3)
plt.xlim(-2,2)
print(len(accepted_x)/N)
```
